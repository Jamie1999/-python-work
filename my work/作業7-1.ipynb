```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 圖表表示中國與美國GDP

```python
b=[2009,2010,2011,2012,2013,2014,2015,2016,2017]
a=[14.42,14.96,15.52,16.16,16.69,17.43,18.12,18.62,19.39]
c=[5.11,6.101,7.573,8.561,9.607,10.48,11.06,11.19,12.24]
```

```python
plt.plot(b,a,label="USA GDP")
plt.plot(b,c,label="China GDP")
plt.title("USA,China 2009~2017 GDP")
plt.xlabel("year")
plt.ylabel("trillion USD")
plt.legend()
plt.show()
```
