# 分析美國主要通訊相關公司股價

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_tmp=pd.read_html("https://tw.stock.yahoo.com/us/s/kimo_listUS7.html?rr=15729529265200.9573216999662257",encoding="big5")
```

```python
len(df_tmp)
```

```python
df=df_tmp[1]
```

```python
df.head()
```

## 刪除掉不必要欄位

```python
df=df.drop("時間",axis=1)
```

```python
df.head()
```

```python
df.describe()
```

## 找出最高股價大於100者

```python
df[df.最高>100].sort_values(by="最高",ascending=False)
```

## 找出最低股價小於10者

```python
df[df.最低<10].sort_values(by="最低",ascending=False)
```

## 由開盤、當日最高排名

```python
df_sorted=df.sort_values(by=["開盤","最高"],ascending=False)
```

```python
df_sorted
```

```python
df_sorted.index=range(1,9)
```

```python
df_sorted
```

## 雙標準篩選資料

```python
df_sorted[(df_sorted.成交>50)|(df_sorted.昨收>50)]
```
